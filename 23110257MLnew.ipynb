```python
#Q2

def matrix_multi(A,B):

    n=len(A)
    k=len(A[0])

    k1=len(B)
    m=len(B[0])
    if k != k1:
        print("No'r of columns in A is not equal to N0'r of rows in B")
    c=[[0 for _ in range (m)]for _ in range (n)]

    for i in range(n):
      for j in range(m):
       for r in range(k):
         c[i][j] += A[i][r]*B[r][j]
    return c

A=np.array([(1,2,3),
    (2,3,4),
    (3,2,1)])
B=np.array([(1,2),
   (3,4),
   (2,6)])
C=matrix_multi(A,B)
for row in C:
    print(row)


A=np.array(A)
B=np.array(B)
C=np.dot(A,B)
print('Results of multiplication of matrices using numpy')
print(C)

# NumPy is fast because it's written in C, which runs much quicker than regular Python code.
```
